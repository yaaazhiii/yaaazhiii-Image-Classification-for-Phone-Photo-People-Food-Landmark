```python
#import google_images_download module
from google_images_download import google_images_download
```

```python
#create ogject
response = google_images_download.googleimagesdownload()

search_queries = ['business individual photography']
```

```python
def downloadimages(query):
    arguments = {'keywords': query, #keywords is the search query
                'format': 'jpg', #image format
                'limit': 100, #the number of images to be downloaded
                'print_urls': True}
    
    try:
        response.download(arguments)
        
    #handling file not found error
    except FileNotFoundError:
        arguments = {"keywords": query, 
                     "format": "jpg", 
                     "limit":100, 
                     "print_urls":True} 
                       
        # Providing arguments for the searched query 
        try: 
            # Downloading the photos based 
            # on the given arguments 
            response.download(arguments)  
        except: 
            pass
        
```

```python
#driver Code
for query in search_queries:
    downloadimages(query)
    print()
```

Standardize Image Format and Size

```python
from PIL import Image
import os.path
import glob

def convertjpg(jpgfile,outdir,width=512,height=512):
    img = Image.open(jpgfile)
    img = img.convert('RGB')
    try:
        new_img = img.resize((width,height),Image.BILINEAR)
        new_img.save(os.path.join(outdir,os.path.basename(jpgfile)))
    except Exception as e:
        print(e)


for jpgfile in glob.glob("C:/Users/Yazhi/Desktop/Test/*.jpg"):
    convertjpg(jpgfile,"C:/Users/Yazhi/Desktop/Test")
    print('Convert Successfully.')
```
